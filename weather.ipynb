````python
# Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Importing the data.

***

- The data needs to be imported using a specific filename format, to make data analysis easier. 
- Using the following format will ensure that the data is store chronololigically rather than numerically.
``` ruby
date +"%Y%m%d_%H%M%S.json"
```

````python
# Dowloading the data from a previously downloaded file
df = pd.read_json('data/weather/20241123_153808_athenry.json')

# Originally I was haaving issues getting the data to load as I read in the data as 'data\weather\20241123_153808_athenry.json', 
# which was causing issues with the backslashes. 
# I had to change the file path to 'data/weather/20241123_153808_athenry.json' to get the data to load correctly.
````

### Analyising the data.

***

````python
# Lets look at the first 3 rows of the data.
print('Table 1: An over of the weather data, Saturday 23rd of November 2024')
df
````

````python
# Lets inspect the variable types.
print('Table 2: The types of data recorded in each of the variables, of the weather data, Saturday 23rd of November 2024.')
df.dtypes
````

````python
# Lets also take a look at a few summary statistics of the data.
print('Table 3: A summary of the weather data, Saturday 23rd of November 2024.')
df.describe()
````

````python
# CoPilot was used to help generate this figure. 
# A number of changes had to be made to the generated code, as it was not working as intended.
# Firstly, it plotted the variable 'date' rather than 'reportTime'.
# Secondly, the code was not present in the figure setting that i wanted to use, so i modified the code to present on a single figure. 

# Create a figure and axis
fig, ax = plt.subplots(5, 1, figsize=(12, 8))

# Plot temperature
ax[0].plot(df.index, df['temperature'], color='tab:red', label='Temperature (C)')
ax[0].set_xlabel('rTime (hr)')
ax[0].set_ylabel('Temperature (C)', color='tab:red')
ax[0].tick_params(axis='y', labelcolor='tab:red')
ax[0].set_title('Temperature (C)', fontsize=12, fontweight='bold', color='orange')

# Plot wind speed
ax[1].plot(df.index, df['windSpeed'], color='tab:blue', label='Wind Speed (kt)')
ax[1].set_xlabel('Time (hr)')
ax[1].set_ylabel('Wind Speed (kt)', color='tab:blue')
ax[1].tick_params(axis='y', labelcolor='tab:blue')
ax[1].set_title('Wind Speed (kt)', fontsize=12, fontweight='bold', color='orange')

# Plot humidity
ax[2].plot(df.index, df['humidity'], color='tab:green', label='Humidity (%)')
ax[2].set_xlabel('Time (hr)')
ax[2].set_ylabel('Humidity (%)', color='tab:green')
ax[2].tick_params(axis='y', labelcolor='tab:green')
ax[2].set_title('Humidity (%)', fontsize=12, fontweight='bold', color='orange')

# Plot rainfall
ax[3].plot(df.index, df['rainfall'], color='tab:purple', label='Rainfall (mm)')
ax[3].set_xlabel('Time (hr)')
ax[3].set_ylabel('Rainfall (mm)', color='tab:purple')
ax[3].tick_params(axis='y', labelcolor='tab:purple')
ax[3].set_title('Rainfall (mm)', fontsize=12, fontweight='bold', color='orange')

# Plot pressure
ax[4].plot(df.index, df['pressure'], color='tab:orange', label='Pressure (mbar)')
ax[4].set_xlabel('Time (hr)')
ax[4].set_ylabel('Pressure (mbar)', color='tab:orange')
ax[4].tick_params(axis='y', labelcolor='tab:orange')
ax[4].set_title('Pressure (mbar)', fontsize=12, fontweight='bold', color='orange')

fig.suptitle ('Figure 1: The Pearsons Correlation Coefficient, the weather data collected, Saturday 23rd of November 2024.', 
             fontsize=14,fontweight='bold', color='orange')
plt.tight_layout()
plt.show()

````

````python
# Lets check if there are correlations between the measured variables.
# Specify the columns to run the correlation on
columns_of_interest = ['temperature', 'windSpeed', 'humidity', 'rainfall', 'pressure']
correlation_matrix = df[columns_of_interest].corr()
correlation_matrix
````

````python
# Generate a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Figure: 1: Correlation Matrix Heatmap.', fontsize=14, fontweight='bold', color='orange')
plt.show()
````
