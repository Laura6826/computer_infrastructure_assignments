## **Task 03** - Formatting Timestamps

### **Assignment Instructions:**

Run the date command again, but this time format the output using YYYYmmdd_HHMMSS (e.g., 20261114_130003 for 1:00:03 PM on November 14, 2026). Refer to the date man page (using man date) for more formatting options. (Press q to exit the man page). Append the formatted output to a file named formatted.txt.

### **Associated Code:**

```ruby
cd data/timestamps                  # This command was used to navigate into the 'timestamps' folder, which is where we want to add additional output.
date +"%Y%m%d_%H%M%S" >> now.txt    # The newly formatted file name was then appended to 'now.txt'.
ls                                  # The 'ls' command was used to verify that the command had run correctly.
                                    # We can clearly see the new file name saved in Table03a.
man date                            # The 'man' command was used to navigate to the 'help' page for the 'date' command, were we would see various formatting options available for the 'date' command.
# 'q' was used to exit the 'man date' page.
date +"%Y%m%d_%H%M%S" >> formatted.txt # Append the formatted output to a file named formatted.txt.
```

1. The following datetime format "%Y%m%d_%H%M%S", will out put a timestamp with the format, year, month, day _ hour, minute, second'.
1. It is essential that there is no space between '+' and '"%Y%m%d_%H%M%S"', otherwise the command will not be recognised.
1. Is it best practice to use this format, with the year first, as it will avoid ording issues, where by files could potentially be organised by the first day of each month, rather than chronologically.
