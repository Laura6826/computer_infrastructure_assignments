## **Task 03** - Formatting Timestamps

### **Assignment Instructions:**

Run the date command again, but this time format the output using YYYYmmdd_HHMMSS (e.g., 20261114_130003 for 1:00:03 PM on November 14, 2026). ### **Associated Code:**

```ruby

                                    # We can clearly see the new file name saved in Table03a.

                                    # command, were we would see various formatting options available for the 
                                    # 'date' command.
                                    #'q' was used to exit the 'man date' page.
```
